```python
#Create an 'X' matrix
X = climate2.iloc[:,:9]
y = answers2.iloc[:,0]
```

```python
window_size = 3

new = np.lib.stride_tricks.sliding_window_view(x=X.values, window_shape=(window_size, X.shape[1]))


reshaped = new.reshape(new.shape[0], window_size, X.shape[1])
print (reshaped)
print (reshaped.shape)
```

```python
y=y.iloc[2:]
y.shape
```

```python
#Split data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(reshaped, y, test_size=0.2, random_state=42)
```

```python
epochs = 30
batch_size = 16
n_hidden = 32

timesteps = window_size
input_dim = len(X_train[0][0])
n_classes = 2

model = Sequential()
model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='sigmoid'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,batch_size=batch_size, epochs=epochs,verbose=1)
acc = model.evaluate(X_train,y_train)
print('Loss:', acc[0], 'Accuracy', acc[1])
```

```python
acc = model.evaluate(X_test,y_test)
print('Loss:', acc[0], 'Accuracy', acc[1])
```

```python
# Create function
def bay_area(neurons, activation, optimizer, learning_rate, batch_size, epochs,
              layers1, layers2, normalization, dropout, dropout_rate): 
    optimizerL = ['SGD', 'Adam', 'RMSprop']
    optimizerD= {'Adam':Adam(learning_rate=learning_rate), 'SGD':SGD(learning_rate=learning_rate),
                 'RMSprop':RMSprop(learning_rate=learning_rate), 'Adadelta':Adadelta(learning_rate=learning_rate),
                 'Adagrad':Adagrad(learning_rate=learning_rate), 'Adamax':Adamax(learning_rate=learning_rate),
                 'Nadam':Nadam(learning_rate=learning_rate), 'Ftrl':Ftrl(learning_rate=learning_rate)}
    activationL = ['relu', 'sigmoid', 'tanh']

    neurons = round(neurons)
    activation = activationL[round(activation)]
    optimizer = optimizerD[optimizerL[round(optimizer)]]
    batch_size = round(batch_size)

    epochs = round(epochs)
    layers1 = round(layers1)
    layers2 = round(layers2)
    
    def lstm_model():
#         epochs = 30 
#         batch_size = 16
#         n_hidden = 32

#         timesteps = window_size
#         input_dim = len(X_train[0][0])
#         n_classes = 2

#         model.add(LSTM(n_hidden, input_shape=(timesteps, input_dim)))
#         model.add(Dropout(0.5))
#         model.add(Dense(n_classes, activation='sigmoid'))
        
        model = Sequential()
        model.add(LSTM(neurons, activation=activation, input_shape=(timesteps, input_dim)))

        if normalization > 0.5:
            model.add(BatchNormalization())
        for i in range(layers1):
            model.add(Dense(neurons, activation=activation)) #(neurons, activation=activation))
        if dropout > 0.5:
            model.add(Dropout(dropout_rate, seed=123))
        for i in range(layers2):
            model.add(Dense(neurons, activation=activation))
        model.add(Dense(n_classes, activation='sigmoid')) #sigmoid softmax
        model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
        return model
    
    es = EarlyStopping(monitor='accuracy', mode='max', verbose=2, patience=20)
    nn = KerasClassifier(build_fn=lstm_model, epochs=epochs, batch_size=batch_size, verbose=2)
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
    score = cross_val_score(nn, X_train, y_train, scoring=score_acc, cv=kfold, fit_params={'callbacks':[es]})
    print (score)
    return score
```

```python
start = time.time()
params ={
    'neurons': (10, 100),
    'activation':(0, 2),
    'optimizer':(0,2),
    'learning_rate':(0.01, 1),
    'batch_size': (200, 1000),
    'epochs':(20, 100),
    'layers1':(1,3),
    'layers2':(1,3),
    'normalization':(0,1),
    'dropout':(0,1),
    'dropout_rate':(0,0.3)
}
# Run Bayesian Optimization
nn_opt = BayesianOptimization(bay_area, params, random_state=42)
nn_opt.maximize(init_points=25, n_iter=25) #25
print('Search took %s minutes' % ((time.time() - start)/60))
```
